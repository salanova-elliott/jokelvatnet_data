# Subsetting JÃ¸kelvatnet data and formatting for plotting


Run code with each type of data desired ("totread", "totrep", and/or "weightrep")

```python
inger_func_raw = pd.read_csv("formatted_data_S1.csv", sep=",", header=0, decimal=",")
data_type = "totrep"
```

```python
# Subsets columnts of interest
info_func_slim = inger_func_raw.loc[:, inger_func_raw.columns.str.contains(f"{data_type}.*EG17_L|{data_type}.*EG17_TS|Taxa_Inger|Functional_group|Category_Inger", regex=True)]
# Drops samples not passing quality filter
info_func_slim.drop(list(info_func_slim.filter(regex = "TS251|L233")), axis = 1, inplace = True)
# Only category 1 taxa
info_cat1 = info_func_slim.loc[inger_func_raw["Category_Inger"] == 1]
# Set indices
info_nona = info_cat1.reset_index().dropna().set_index(["Functional_group", "Category_Inger", "Taxa_Inger"])
# Drop old useless numbered index
info_nona.drop("index", inplace=True, axis=1)
# Transpose
info_t = info_nona.T
# Collapses identical taxa
if data_type == "totread":
    info_u = info_t.groupby(info_t.columns, axis=1).sum()
elif data_type == "totrep":
    info_u = info_t.groupby(info_t.columns, axis=1).max()
else:
    raise ValueError()
#info_u = info_t.loc[:, ~info_t.columns.duplicated()] not needed since above code collapses
# Removes taxa that do not appear in the dataset
info_slimmer = info_u.loc[:, (info_u != 0).any(axis=0)]
info_slimmer
```

```python
# Turns tuples back into mutli-index
info_slimmer.columns = pd.MultiIndex.from_tuples(info_slimmer)
info_slimmer
```

```python
info_slimmer.drop(("Tree", 1, "Pinus"), axis=1, inplace=True)
info_slimmer.drop(("Algae", 1, "Cosmarium botrytis"), axis=1, inplace=True)
info_slimmer.drop(("Algae", 1, "Nannochloropsis"), axis=1, inplace=True)
info_slimmer
```

```python
# Orders rows based on age (TS samples were originally all last)
inger_slimmer = info_slimmer.reindex([f"{data_type}_sample.EG17_L001_rpt",
f"{data_type}_sample.EG17_L006_rpt",
f"{data_type}_sample.EG17_L012_rpt",
f"{data_type}_sample.EG17_L018_rpt",
f"{data_type}_sample.EG17_TS025_rpt",
f"{data_type}_sample.EG17_L031_rpt",
f"{data_type}_sample.EG17_L038_rpt",
f"{data_type}_sample.EG17_L045_rpt",
f"{data_type}_sample.EG17_L052_rpt",
f"{data_type}_sample.EG17_L059_rpt",
f"{data_type}_sample.EG17_L066_rpt",
f"{data_type}_sample.EG17_L073_rpt",
f"{data_type}_sample.EG17_L080_rpt",
f"{data_type}_sample.EG17_L087_rpt",
f"{data_type}_sample.EG17_L094_rpt",
f"{data_type}_sample.EG17_L101_rpt",
f"{data_type}_sample.EG17_L109_rpt",
f"{data_type}_sample.EG17_L116_rpt",
f"{data_type}_sample.EG17_L123_rpt",
f"{data_type}_sample.EG17_L130_rpt",
f"{data_type}_sample.EG17_L137_rpt",
f"{data_type}_sample.EG17_L144_rpt",
f"{data_type}_sample.EG17_TS151_rpt",
f"{data_type}_sample.EG17_L159_rpt",
f"{data_type}_sample.EG17_L166_rpt",
f"{data_type}_sample.EG17_L173_rpt",
f"{data_type}_sample.EG17_L179_rpt",
f"{data_type}_sample.EG17_L185_rpt",
f"{data_type}_sample.EG17_L191_rpt",
f"{data_type}_sample.EG17_L197_rpt",
f"{data_type}_sample.EG17_L203_rpt",
f"{data_type}_sample.EG17_L209_rpt",
f"{data_type}_sample.EG17_L215_rpt",
f"{data_type}_sample.EG17_L221_rpt",
f"{data_type}_sample.EG17_L227_rpt",
f"{data_type}_sample.EG17_L239_rpt",
f"{data_type}_sample.EG17_L245_rpt",
f"{data_type}_sample.EG17_L257_rpt"])
```

```python
# Writes files for each functional group
for taxa in ["Bryophyte","Clubmoss", "Horsetail",
             "Fern", "Forb", "Graminoids", "Tree", "Dwarf_shrub", "Shrub"]:
    new_slice = inger_slimmer[taxa]
    new_slice.columns = new_slice.columns.droplevel()
    new_slice.to_csv(f"subsets_{data_type}/{taxa}.tsv", sep="\t", index=False)
```

```python
# Writes a file containing all taxa
only_taxa_headers = inger_slimmer.copy(deep=True)
only_taxa_headers.columns = only_taxa_headers.columns.droplevel()
only_taxa_headers.columns = only_taxa_headers.columns.droplevel()
only_taxa_headers.to_csv(f"subsets_{data_type}/all_taxa.tsv", sep="\t", index=False)
#inger_slimmest.to_csv("subsets_totread/all_slim4_taxa.tsv", sep="\t", index=False)
```

```python
# Sums functional group counts
group_sums = pd.DataFrame()

for taxa in ["Bryophyte","Clubmoss", "Horsetail", "Shrub",
             "Fern", "Forb", "Graminoids", "Tree", "Dwarf_shrub"]:
    group_sums[taxa] =  inger_slimmer[taxa].sum(axis=1)

group_sums["Saliceae"] = inger_slimmer[("Tree",1, "Saliceae")]
group_sums["Woody_taxa"] = group_sums["Tree"] - group_sums["Saliceae"] + group_sums["Shrub"]
group_sums["Vascular_cryptogam"] = group_sums["Fern"] + group_sums["Clubmoss"] + group_sums["Horsetail"]
group_sums = group_sums.drop("Tree", axis=1)
group_sums = group_sums.drop("Shrub", axis=1)
group_sums = group_sums.drop("Fern", axis=1)
group_sums = group_sums.drop("Clubmoss", axis=1)
group_sums = group_sums.drop("Horsetail", axis=1)

group_percent = group_sums.div(group_sums.sum(axis=1), axis=0)
group_percent
```

```python
percent_rows = inger_slimmer.div(inger_slimmer.sum(axis=1), axis=0)
percent_rows
```

```python
for taxa in ["Bryophyte","Clubmoss", "Horsetail",
             "Fern", "Forb", "Graminoids", "Tree", "Dwarf_shrub", "Shrub"]:
    new_slice = percent_rows[taxa]
    new_slice.columns = new_slice.columns.droplevel()
    new_slice.to_csv(f"subsets_{data_type}/{taxa}_percent.tsv", sep="\t", index=False)
```
